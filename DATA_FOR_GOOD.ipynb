```python
# Zero down to the RESPONSE variable:

# Remove all the N/A values from the dataframe:
data_2020_numeric.dropna(how='all')
```

```python
# Drop N/A values across rows:

data_2020_COVID = data_2020
```

```python
# Create the required X variable:
X = data_2020_numeric.drop(['RespondentID','newid','collectorid','hlth_happy'], axis =1 )

y = data_2020_numeric['hlth_happy']

X = np.nan_to_num(X)

y = np.nan_to_num(y)
```

```python
print(len(X))
print(len(y))
```

```python
# Split the dataframe into test and train:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Implement the Random Forest Classfier:

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

clf = RandomForestClassifier(random_state=0)
```

```python
# Fit the required Random Forest:

clf.fit(X_train, y_train)
```

```python
# Predict the values with trained model:

y_pred = clf.predict(X_test)
```

```python
# Evaluate the model:

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Look for the features that are important:

from sklearn.feature_selection import SelectFromModel
sfm = SelectFromModel(clf, threshold=0.05)
```

```python
sfm

# Train the selector
sfm.fit(X_train, y_train)

# Print the name of these important features:
for feature_list_index in sfm.get_support(indices=True):
    print(data_2020_numeric[feature_list_index])
```
