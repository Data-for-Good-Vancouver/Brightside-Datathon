```python
# Import the required libraries:
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data_2020 = pd.read_csv("C:/Users/mural/Desktop/data_for_good/2020_brightside_data.csv")
data_2020.head(3)
```

```python
data_2020.describe()
```

```python
data_2020.info()
```

```python
# Get the numeric data:

data_2020_numeric = data_2020._get_numeric_data()
data_2020_numeric.head()
```

```python
# Create the required X and Y variable:

X = data_2020_numeric.drop(['RespondentID','newid','collectorid','hlth_happy'], axis =1 )

y = data_2020_numeric['hlth_happy']

X = np.nan_to_num(X)

y = np.nan_to_num(y)
```

```python
print(len(X))
print(len(y))
```

```python
# Split the dataframe into test and train:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Implement the Random Forest Classfier:

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

clf = RandomForestClassifier(random_state=0)
```

```python
# Fit the required Random Forest:

clf.fit(X_train, y_train)
```

```python
# Predict the values with trained model:

y_pred = clf.predict(X_test)
```

```python
# Evaluate the model:

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Look for the features that are important:

from sklearn.feature_selection import SelectFromModel
sfm = SelectFromModel(clf, threshold=0.1)

# Train the selector
sfm.fit(X_train, y_train)

# Print the name of these important features:
for feature_list_index in sfm.get_support(indices=True):
    print(data_2020_numeric[feature_list_index])
```

```python
print(sfm.get_support(indices = True))
```

```python
# Find the importance factors of the features:
clf.feature_importances_
```

```python
# Find the features which are the most important:
cor = data_2020_numeric.corr()
#sns.heatmap(cor, annot=True)
cor.head()
```

```python
#Correlation with output variable
cor_target = abs(cor["hlth_happy"])

#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.25]
relevant_features
```

### This is a pretty big insight for determining what are the variables that affect the final happiness ratings of the residents
